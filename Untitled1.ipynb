```python
%matplotlib inline
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import matplotlib
df = pd.read_csv('C:/Work/2008.csv')
df.head(5)
```

```python
del df['CancellationCode']
```

```python
df['DepTime'] = pd.to_datetime(df.DepTime, unit='m').dt.strftime('%H:%M')
df['ArrTime'] = pd.to_datetime(df.ArrTime, unit='m').dt.strftime('%H:%M')
df['CRSDepTime'] = pd.to_datetime(df.CRSDepTime, unit='m').dt.strftime('%H:%M')
df['CRSArrTime'] = pd.to_datetime(df.CRSArrTime, unit='m').dt.strftime('%H:%M')
```

```python
df1 = df.dropna()
df1.head(5)
```

```python
df1.describe()
```

```python
def dict_sort(my_dict):
    keys = []
    values = []
    my_dict = sorted(my_dict.items(), key=lambda x:x[1], reverse=True)
    for k, v in my_dict:
        keys.append(k)
        values.append(v)
    return (keys,values)
def getColors(n):
    COLORS = []
    cm = plt.cm.get_cmap('hsv', n)
    for i in np.arange(n):
        COLORS.append(cm(i))
    return COLORS
```

```python
month_count = pd.value_counts(df1['Month'].values, sort=True)
month_count_keys, month_count_values = dict_sort(dict(month_count))    
TOP_MONTH = len(month_count_keys)
plt.title('Наблюдения по месяцам', fontsize=14)
plt.bar(np.arange(TOP_MONTH), month_count_values, color=getColors(TOP_MONTH))
plt.xticks(np.arange(TOP_MONTH), month_count_keys, rotation=0, fontsize=8)
plt.yticks(fontsize=14)
plt.ylabel('Количество наблюдений', fontsize=14)
plt.xlabel('Месяц', fontsize=14)
plt.show()
```

```python
air_count = pd.value_counts(df1['UniqueCarrier'].values, sort=True)
air_count_keys, air_count_values = dict_sort(dict(air_count))    
TOP_COUNTRY = len(air_count_keys)
plt.title('Аэропорты, где больше всего наблюдений', fontsize=14)
plt.bar(np.arange(TOP_COUNTRY), air_count_values, color=getColors(TOP_COUNTRY))
plt.xticks(np.arange(TOP_COUNTRY), air_count_keys, rotation=0, fontsize=8)
plt.yticks(fontsize=14)
plt.ylabel('Количество наблюдений', fontsize=14)
plt.xlabel('Код аэропорта', fontsize=14)
plt.show()
```

### 5.Проверить гипотезы с помощью эконометрики/машинного обучения. 

```python
dummies = pd.get_dummies(df1['DayOfWeek'])
sat_sun = dummies[7]+dummies[6]
df1.insert(1, "Sat_Sun", sat_sun)
```

```python
from sklearn.linear_model import LinearRegression
X1 = df1.Distance.values.reshape(-1, 1)
y1 = df1.Sat_Sun.values
reg = LinearRegression().fit(X1, y1)
reg.score(X1, y1)
```

```python
import statsmodels.api as sm
from scipy import stats
X2 = sm.add_constant(X1)
est = sm.OLS(y1, X2)
est2 = est.fit()
print(est2.summary())
```
